```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_train = np.genfromtxt('lab_1_train.csv', delimiter = ',')
file_test = np.genfromtxt('lab_1_test.csv', delimiter = ',')

X_train = file_train[1:,1]
Y_train = file_train[1:,2]

X_test = file_test[1:,1]
Y_test = file_test[1:,2]

X_train.shape
Y_train.shape

X_test.shape
Y_test.shape
```

```python
plt.scatter(X_train, Y_train)
plt.scatter(X_test, Y_test)
```

```python
def calc_cost(y_pred, y):
    return 0.5 * np.sum((y_pred - y)**2)
```

```python
def grad_w(y_pred, y, x):
    return np.sum((y_pred - y) * x)

def grad_b(y_pred, y):
    return np.sum(y_pred - y)
```

```python
def gradient_descent(X_train,Y_train, w, b, learning_rate, eps, stop = 1000):
    y_pred = X_train * w + b
    
    for i in range(stop) :
        last_pred = y_pred
        w = w - learning_rate * grad_w(y_pred, Y_train, X_train)
        b = b - learning_rate * grad_b(y_pred, Y_train)
        y_pred = X_train * w + b
        delta = np.linalg.norm(last_pred-y_pred)
        cost = calc_cost(y_pred, Y_train)
        print(f'{i} cost {cost}\tw {w}\tb {b}')
        if delta < eps:
            break
    return w, b
```

```python
def model(X_train, X_test, Y_train, Y_test, learning_rate, eps):
    w = 0.0
    b = 0.0
    w, b = gradient_descent(X_train, Y_train, w, b, learning_rate, eps)
    Y_prediction_train = X_train * w + b
    Y_prediction_test = X_test * w + b
    cost_train = calc_cost(Y_prediction_train, Y_train)
    cost_test = calc_cost(Y_prediction_test, Y_test)
    print("train cost = ", cost_train, "test cost = ", cost_test)
    return w, b
    
```

```python
w, b = model(X_train, X_test, Y_train, Y_test, 0.015, 0.001)
```

```python
plt.scatter(X_train, Y_train)
plt.scatter(X_test, Y_test)
x = np.linspace(0, 1, 10)
plt.plot(x, w*x+b)
```
